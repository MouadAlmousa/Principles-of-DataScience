```python
pd.options.display.max_columns = None
pd.options.display.expand_frame_repr =False
```

Read in the clean data using pandas

```python
df = pd.read_csv('../data_clean/clean_yield_data.csv')

```

Get Name, Location and Price for Diesel cars sort by prices(ascending)

```python
df_Diesel = df.loc[df['Fuel_Type_Diesel'] == True, ['Name', "Location",'Price']].sort_values(by='Price')
df_Diesel.head(5)
```

Find average Price,Mileage and Power grouped by Year

```python
df_summary = df.groupby('Year').agg(
    Avg_price=('Price', 'mean'),
    Avg_Mileage=('Mileage_kmpl', 'mean'),
    Avg_Power=('Power_bhp', 'mean'),
    Car_Count=('Name', 'count')
).round(3)
df_summary.head(20)
```
